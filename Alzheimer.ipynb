

**Mixed Correlation Matrix**

Because it combines different statistical correlation methods based on the data type:

Data Type Pair	               Correlation Used	                     Meaning  
Numeric ↔ Numeric	          Pearson correlation	      Measures linear relationship  
Categorical ↔ Categorical	  Cramér’s V	              Measures association between categories  
Numeric ↔ Categorical	      Correlation Ratio (η)	      Measures how well categories explain       numeric values





```python

import scipy.stats as ss


# ------------------------
# Helpers
# ------------------------

# Cramér's V (categorical vs categorical)
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r, k = confusion_matrix.shape
    phi2_corr = max(0, phi2 - (k-1)*(r-1)/(n-1))
    r_corr = r - (r-1)**2/(n-1)
    k_corr = k - (k-1)**2/(n-1)
    return np.sqrt(phi2_corr / min((k_corr-1), (r_corr-1)))

# Correlation Ratio η (categorical vs numerical)
def correlation_ratio(categories, values):
    fcat, _ = pd.factorize(categories)
    cat_num = np.max(fcat) + 1
    y_avg = np.mean(values)

    numerator = 0
    for i in range(cat_num):
        cat_values = values[fcat == i]
        if len(cat_values) == 0:
            continue
        numerator += len(cat_values) * (np.mean(cat_values) - y_avg)**2
        
    denominator = np.sum((values - y_avg)**2)
    if denominator == 0:
        return 0
    return np.sqrt(numerator / denominator)

# ------------------------
# Build MIXED correlation matrix
# ------------------------

def mixed_corr(df):
    cols = df.columns
    size = len(cols)

    corr_matrix = pd.DataFrame(np.zeros((size, size)), columns=cols, index=cols)

    for i in range(size):
        for j in range(size):
            col_i = df[cols[i]]
            col_j = df[cols[j]]

            # numeric vs numeric → Pearson
            if col_i.dtype != 'object' and col_j.dtype != 'object':
                corr = col_i.corr(col_j)

            # categorical vs categorical → Cramér’s V
            elif col_i.dtype == 'object' and col_j.dtype == 'object':
                corr = cramers_v(col_i, col_j)

            # numeric vs categorical → Correlation Ratio
            elif col_i.dtype != 'object' and col_j.dtype == 'object':
                corr = correlation_ratio(col_j, col_i)

            else:  # categorical vs numeric
                corr = correlation_ratio(col_i, col_j)

            corr_matrix.iloc[i, j] = corr

    return corr_matrix

# ------------------------
# RUN IT
# ------------------------

mixed_corr_matrix = mixed_corr(data)

plt.figure(figsize=(30, 18))
sns.heatmap(mixed_corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Mixed-Type Correlation Matrix (Numeric + Categorical)", fontsize=18)
plt.show()
```

## **HANDLING CLASS IMBALANCE USING SMOTE**  
SMOTE (Synthetic Minority Oversampling Technique)

SMOTE is a popular technique used to handle class imbalance in machine learning datasets. When one class (usually the positive/minority class) has far fewer examples than the other, standard models tend to ignore the minority class, leading to poor recall and F1 scores. SMOTE addresses this problem by creating synthetic samples for the minority class instead of simply duplicating existing ones.

```python

# use SMOTE to handle class imbalance
from imblearn.over_sampling import SMOTE

# X_train and y_train should be your full training data
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train_final, y_train)

print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", y_train_res.value_counts())
```

Before SMOTE, the dataset was imbalanced with 1111 samples of class 0 and 608 samples of class 1.
This imbalance can cause models to favor the majority class, leading to poor detection of the minority class.
SMOTE generates synthetic samples for the minority class to balance the dataset.
After SMOTE, both classes have 1111 samples, making the dataset perfectly balanced.
Training on this balanced data should improve the model’s recall and F1-score for the minority class.

## **RETRAINING CATBOOST WITH CLASS WEIGHTS AND UPDATED HYPERPARAMETERS**

Class weights are in the hyperparameter to automatically handle class imbalance.

```python

cat_features_new = X_train_final.select_dtypes(include=['category']).columns.tolist()
print("Categorical features:", cat_features_new)

# Parameter grid for CatBoost
param_griddy = {
    'iterations': [500, 800, 1000, 1500], 
    'learning_rate': [0.01, 0.03, 0.05, 0.1],
    'depth': [4, 6, 8, 10],
    'l2_leaf_reg': [1, 3, 5, 7, 9],
    'bagging_temperature': [0.0, 0.5, 1.0],
    'class_weights': [[1, 1], [1, 2], [1, 3]],  # different class weightings
    'random_seed': [42, 4, 7, 21]    
}

kfoy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

modeler = CatBoostClassifier(
    verbose=0,
    cat_features=cat_features_new
)

random_searcher = RandomizedSearchCV(
    estimator=modeler,
    param_distributions=param_griddy,
    n_iter=20,
    scoring='roc_auc',
    cv=kfoy,
    random_state=42,
    n_jobs=-1
)

random_searcher.fit(X_train_final, y_train)

print("Best Parameters:", random_searcher.best_params_)
print("Best CV AUC:", random_searcher.best_score_)
```

```python
best_modeller = CatBoostClassifier(
    **random_searcher.best_params_,
    verbose=0,
    cat_features=cat_features_new
)

best_modeller.fit(X_train_final, y_train)
```

**EXPLANATION: THE LARGER THE AREA UNDER THE CURVE,  THE BETTER HOW THE MODEL IS**

The ROC curve is a graphical representation of a classifier’s performance across all possible classification thresholds. It shows the trade-off between the True Positive Rate (TPR) and the False Positive Rate (FPR).

True Positive Rate (TPR) / Sensitivity / Recall:
Measures the proportion of actual positives correctly identified.

**TPR = TP/(TP + FN)**

False Positive Rate (FPR):
Measures the proportion of actual negatives incorrectly classified as positive.

**FPR = FP/(FP + TN)**

**FPR = 1 - Specificity (True Negative Rate )**

**SPECIFICITY/TRUE NEGATIVE RATE = TN/(TN + FP)**

## **FINAL FEATURE IMPORTANCE ANALYSIS AND VISUALIZATION**

```python

from catboost import Pool

# --- Step 1: Get feature importance ---
feature_importances = best_modeller.get_feature_importance(
    Pool(X_train_final, y_train, cat_features=cat_features_new) # The Pool tells catboost, here’s the dataset with labels, and these columns are categorical.”
)
feature_names = X_train_final.columns

# --- Step 2: Create a DataFrame ---
fi_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importances
})

# --- Step 3: Sort by importance (descending) ---
fi_df = fi_df.sort_values(by='Importance', ascending=False).reset_index(drop=True)
fi_df['Rank'] = fi_df.index + 1  # add rank column

# --- Step 4: Plot horizontal bar chart ---
plt.figure(figsize=(12,9))
bars = plt.barh(fi_df['Feature'], fi_df['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.title('Feature Importance (Ranked)')
plt.gca().invert_yaxis()  # highest importance on top

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.1, bar.get_y() + bar.get_height()/2,
             f'{width:.2f}', va='center', fontsize=10)
    
plt.show()



# --- Step 5: Show table ---
print("\nFeature Importance Table (Ranked)")
print(fi_df[['Rank', 'Feature', 'Importance']])
```

CatBoost measures feature importance by seeing how much each feature helps reduce prediction error when used to split nodes in the trees.
Features that cause larger improvements in predictions across all trees are ranked higher.  

The model checks how much each feature helps in reducing prediction errors across all trees. Features that split the data in ways that most reduce mistakes are ranked higher in importance.”

“CatBoost looks at how much each feature helps the model make better predictions, and gives higher scores to features that are more useful.”
