```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
vgsales['Year'] = imputer.fit_transform(vgsales['Year'].values.reshape(-1, 1))
vgsales.isna().sum()
```

DATA IMPUTATION

```python
imputer = SimpleImputer(strategy='most_frequent')
vgsales[['Publisher']] = imputer.fit_transform(vgsales[['Publisher']])
vgsales.isna().sum()
```

CONVERTING CATEGORICAL DATA INTO NUMERICAL DATA

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
vgsales['Platform'] = le.fit_transform(vgsales['Platform'])
vgsales['Genre'] = le.fit_transform(vgsales['Genre'])
vgsales['Publisher'] = le.fit_transform(vgsales['Publisher'])
vgsales['Name'] =le.fit_transform(vgsales['Name'])
vgsales
```

```python

X = vgsales.drop('Global_Sales', axis=1)
Y = vgsales['Global_Sales']

```

FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
Y = sc.fit_transform(Y.values.reshape(-1, 1))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

INSTANTIATING THE MODEL

```python
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(random_state=42),
    "XGBoost": XGBRegressor(random_state=42, verbosity=0),
    "CatBoost": CatBoostRegressor(verbose=0, random_state=42)
}
```

PERFORMANCE METRICS

```python

from sklearn.metrics import mean_squared_error, r2_score
results = []
for name, model in models.items():
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)

    mse = mean_squared_error(Y_test, y_pred)
    rmse = mse ** 0.5
    r2 = r2_score(Y_test, y_pred)
    results.append({"Model": name, "RMSE": rmse, "R2 Score": r2, "MSE": mse})
```

```python
results
```

1. Linear Regression

RMSE = 0.0033 → On average, your predictions are off by only 0.003 sales units (almost nothing!).

MSE = 1.12e-05 → Errors are extremely tiny.

R² = 0.99999 → The model explains ~99.999% of the variance in your data.

⚠️ Interpretation: This performance is too good to be true. Likely, Linear Regression is overfitting because the data is simple, collinear, or your features leak information about the target. It’s suspiciously perfect.

2. Random Forest

RMSE = 0.5286 → Predictions are off by about 0.53 sales units on average.

MSE = 0.2794 → Moderate squared errors.

R² = 0.839 → Explains about 84% of the variance in global sales.

✅ More realistic performance. Shows that it generalizes decently but isn’t perfect.

3. XGBoost

RMSE = 0.5995 → Errors are ~0.60 sales units.

MSE = 0.3594.

R² = 0.793 → Explains ~79% of the variance.

⚠️ Slightly worse than Random Forest. Could be due to needing better hyperparameter tuning.
